```python
# Importing relevant packages to carry out data analysis and create visulisations
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import numpy as np
from datetime import timedelta
```

```python
### General functions ###

# Function to export dataframe as a markdown table to copy into HackMD
def export_df_as_markdown(dataframe, name):
    markdown_table = dataframe.to_markdown(index=False)
    with open(f'{name}_markdown_table.txt', 'w', encoding='utf-8') as file:
        file.write(markdown_table)
```

```python
# Read the Combined Data CSV file and create a dataframe
try:
    df = pd.read_csv('marathon_combined_data.csv')

# Error handling
except FileNotFoundError:
    print("Error: File not found.")
except Exception as e:
    print(f"An error occurred: {e}")

# Convert 'Year' and 'Position' columns to integer values
df['Year'] = df['Year'].astype(int)
df['Position'] = df['Position'].astype(int)

# Showing the 54 year marathon, creating markdown table of line and 
fifty_four_year_marathon = df[(df['Name'] == 'Shizo Kanakuri') & (df['Year'] == 1912)]
export_df_as_markdown(fifty_four_year_marathon, '54_year_marathon')
display(fifty_four_year_marathon)

# Removing 54 year marathon from dataframe as it is a large anomaly in the data that could skew analysis
df = df.drop(fifty_four_year_marathon.index[0]).reset_index(drop=True)

# Adding OR (Olympic Record) for first Women Marathon finisher
first_women_finisher_index = df[(df['Year'] == 1984) & (df['Gender'] == 'Women') & (df['Position'] == 1)].index[0]
df.loc[first_women_finisher_index, 'Notes'] = 'OR'
display(df.loc[first_women_finisher_index])

# Removing notes other than world record (WR) and olympic record (OR) and replacing with '-'
for row in range(len(df)):
    if df.loc[row, 'Notes'] not in ['WR', 'OR']:
        df.loc[row, 'Notes'] = '-'

# Displaying the dataframe
display(df)
```

```python
# Function to convert the time string into a time delta format
def time_str_to_time_delta(time_str):
    # Splitting the string into 3 parts by their separation by a colon
    parts = time_str.split(':')
    # Assigning the first part to the hours
    hours = int(parts[0])
    # Assigning the second part to the minutes
    minutes = int(parts[1])
    # Setting microseconds to zero for if there is no microseconds in the string
    microseconds = 0
    if '.' in parts[2]:
        seconds, tenths = map(int, parts[2].split('.'))
        microseconds = tenths * 100000
    else:
        seconds = int(parts[2])
    return timedelta(hours=hours, minutes=minutes, seconds=seconds, microseconds=microseconds)

df['Time'] = df['Time_Taken'].apply(time_str_to_time_delta)

def time_difference(time_1, time_2):
    if time_2 < time_1:
        time_1, time_2 = time_2, time_1
    # Calculate the difference
    difference = time_2 - time_1
    # Print the difference
    return difference

def time_difference_to_words(difference):
    minutes = difference.seconds // 60
    seconds = difference.seconds % 60
    hundredths = difference.microseconds // 10000
    return f'Difference: {minutes} minutes, {seconds} seconds, {hundreths} hundreths of second'



```

```python
def format_y_axis_times_from_minutes():
    # Get the current y-axis tick labels
    tick_labels = plt.gca().get_yticks()
    # Convert tick labels from float hours to "HH:MM" format
    tick_labels_formatted = [f"{int(minutes) // 60:02d}:{int(minutes) % 60:02d}" for minutes in tick_labels]    # Set the modified tick labels back to the y-axis
    plt.gca().set_yticklabels(tick_labels_formatted)

def format_y_axis_times():
    # Get the current y-axis tick labels
    tick_labels = plt.gca().get_yticks()
    # Convert tick labels from float hours to "HH:MM" format
    tick_labels_formatted = [f"{int(seconds) // 3600:02d}:{int(seconds) % 3600 // 60:02d}" for seconds in tick_labels]
    plt.gca().set_yticklabels(tick_labels_formatted)
```

```python
### Total Finishers by Gender Each Year ###
plt.style.use('default')
plt.grid(True, alpha=0.5, linestyle=":")

for gender in ['Men', 'Women']:
    grouped_counts = df[df['Gender'] == gender].groupby('Year').size()
    # Calculate the number of bars (assuming each bar represents a year)
    num_years = len(grouped_counts)
    #print(num_years)

    # Calculate the figure size based on the number of bars
    fig_width = (num_years // 2)  # Minimum width of 8 inches, adjust multiplier as needed
    fig_height = 6  # You can adjust this as needed
    plt.figure(figsize=(fig_width, fig_height))
    
    # Plot the grouped counts
    ax = grouped_counts.plot(kind='bar', color=['#85e085'], width=0.5)

    # Add labels and title
    plt.xlabel('Year')
    plt.ylabel('Number of finishers')
    plt.title(f'Finishers Each Year ({gender})')
    # Annotate each bar with its count value
    for i, count in enumerate(grouped_counts):
        ax.annotate(str(count), xy=(i, count), ha='center', va='bottom')
        
    # Save plot as PNG
    plt.savefig(f'total_finishers_{gender.lower()}.png', bbox_inches='tight')

    # Show plot
plt.show()
```

```python
men_and_women_winners = winners_df[winners_df['Year']>=1984].sort_values(by='Year')
display(men_and_women_winners)

men_and_women_averages = df[df['Year']>=1984].groupby(['Gender', 'Year'])['Time'].mean().reset_index()
men_and_women_averages = men_and_women_averages.sort_values(by=['Year', 'Gender']).reset_index(drop=True)

# Create new columns 'time_delta' and 'difference'
men_and_women_averages['time_delta'] = None
men_and_women_averages['difference'] = None
men_and_women_averages['percentage_difference'] = None

previous_time = None
# Assume df is your DataFrame
for index, row in men_and_women_averages.iterrows():
    # Accessing a specific column value for each row
    current_time = row['Time']
    
    # Accessing the value in the row above
    if previous_time is not None:
        # Update the previous value for the next iteration
        difference = time_difference(current_time, previous_time)
        # Update the DataFrame with the calculated values
        if row['Gender'] == 'Men':
            men_and_women_averages.at[index, 'time_delta'] = None
            men_and_women_averages.at[index, 'difference'] = None
            men_and_women_averages.at[index, 'percentage_difference'] = None
            
        else:
            men_and_women_averages.at[index, 'time_delta'] = difference
            men_and_women_averages.at[index, 'difference'] = time_difference_to_words(difference)
            men_and_women_averages.at[index, 'percentage_difference'] = (difference/previous_time)*100
    previous_time = current_time
display(men_and_women_averages)

women_data = men_and_women_averages[men_and_women_averages['Gender'] == 'Women']
# Convert 'Time' to total seconds for plotting
women_data['Time_seconds'] = [pd.Timedelta(time).total_seconds() for time in women_data['time_delta']]

# Plotting
plt.figure(figsize=(10, 6))

# Plot Women
plt.scatter(women_data['Year'], women_data['percentage_difference'], label='Women', marker='o', color='green')
# Annotate each point with the percentage difference
for i, row in women_data.iterrows():
    plt.annotate(f"{row['percentage_difference']:.2f}%", 
                 (row['Year'], row['percentage_difference']),
                 textcoords="offset points",
                 xytext=(25,-3), 
                 ha='center')

women_data['percentage_difference'] = pd.to_numeric(women_data['percentage_difference'], errors='coerce')

# Fit a linear regression line
coefficients = np.polyfit(women_data['Year'], women_data['percentage_difference'], 1)
poly = np.poly1d(coefficients)
x_values = np.linspace(min(women_data['Year']), max(women_data['Year']), 100)
y_values = poly(x_values)

# Plot the line of best fit
plt.plot(x_values, y_values, color='orange', linestyle='--', label='Line of Best Fit')

# Labels and title
plt.xlabel('Year')
plt.ylabel('Time (seconds)')
plt.title('Differences in Winning Times Over Years')
plt.legend()
plt.grid(True, alpha=0.5, linestyle=":")
plt.xticks(rotation='vertical')

# Show plot
plt.tight_layout()
plt.show()
```
